```python
import pandas as pd
from sqlalchemy import create_engine
 
#load your cleaned CSV
df = pd.read_csv("london_prescribing_clean.csv")
 
#connect to MySQL
engine = create_engine("mysql+pymysql://root:root@localhost/nhs_project")
 
#this creates/replaces the table automatically
df.to_sql("london_prescribing", con=engine, if_exists="replace", index=False, chunksize=5000)
 
print("Done! Data loaded into MySQL.")
print(f"Rows loaded: {len(df)}")
```
